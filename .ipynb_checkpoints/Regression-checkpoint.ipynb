# Linear regression w.r. to OLS(Ordinary Least Squares)

Make linear regression class with gradient descent optimization from scratch.

```python
class LinearRegressionGD:
    def __init__(self, eta=0.01, n_iter=50, random_state=1, J_type='MSE'):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state
        self.J_type = J_type

    def fit(self, X, y):
        """
        fits the model for training data
        """
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=X.shape[1]) # initialize weights
        self.b_ = np.array([0.]) # intercept
        self.losses_ = [] # errors

        for i in range(self.n_iter):
            output = self.net_input(X)
            errors = (y - output)
            if self.J_type == 'MSE':
                self.w_ += self.eta * 2.0 * X.T @ errors / X.shape[0] # update weights
                self.b_ += self.eta * 2.0 * errors.mean() # update intercept
                loss = (errors**2).mean() # objective function J(w)
            elif self.J_type == 'SSE':
                self.w_ += self.eta * X.T @ errors  # update weights
                self.b_ += self.eta * errors.sum() # update intercept
                loss = (errors**2).sum()/2.0 # objective function J(w)
                
            self.losses_.append(loss)
        return self

    def net_input(self, X):
        """
        returns value for target y
        """ 
        return X @ self.w_ + self.b_

    def predict(self, X):
        """
        makes prediction on given test/unseen data set
        """ 
        return self.net_input(X)
```

We set 'Overall Qual' as an explanatory variable for 'SalePrice'

```python
#X = df[['Overall Qual']].values
X = df[['Gr Liv Area']].values
y = df['SalePrice'].values
```

Scale data

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()

X_std = sc_x.fit_transform(X)
y_std = sc_y.fit_transform(y[:,np.newaxis]).flatten()
# Since standard scaler takes multidim arrays we use np.newaxis.
# np.newaxis adds a new dimension, then it is flattened by flatten() 
# to get ordinary column vector shape
```

```python
# model
lr = LinearRegressionGD(eta=0.1,J_type='MSE')
lr.fit(X_std, y_std)
```

```python
plt.plot(range(1, lr.n_iter+1), lr.losses_)
plt.ylabel(lr.J_type)
plt.xlabel('Epoch')
plt.tight_layout()
plt.show()
```

### Visualize results

```python
def lin_regr_plot(X, y, model):
    plt.scatter(X,y, c='steelblue', s=70)
    plt.plot(X, model.predict(X), color='red', lw=2)
    plt.xlabel('Scaled Living area above ground')
    plt.ylabel('Scaled Sale Price')
    return None
```

```python
lin_regr_plot(X_std, y_std, lr)
```

```python
feature_std = sc_x.transform(np.array([[2500]]))
target_std = lr.predict(feature_std)
target_reverted = sc_y.inverse_transform(target_std.reshape(-1, 1))
print(f'Sale price: ${target_reverted.flatten()[0]:.2f}')
```

```python
print(f'Slope: {lr.w_[0]:.3f}')
print(f'Intercept: {lr.b_[0]:.3f}') # when data is scaled, the intercept is always 0
```

# Analytical solution via Normal Equation

$\vec{w} = (X^TX)^{-1} X^{T}\vec{y}$

```python
# standardized features and data 
Xb = np.hstack((np.ones((X.shape[0],1)), X_std)) # features matrix
w = np.zeros(X_std.shape[1]) # weights
z = np.linalg.inv(Xb.T @ Xb) # inverse matrix
w = z @ Xb.T @ y_std

print(f'Slope: {w[1]:.3f}')
print(f'Intercept {w[0]:.3f}')
```

# Estimate coefficients via sklearn

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr.fit(X_std, y_std)
y_pred = slr.predict(X_std)


print(f'Slope: {slr.coef_[0]:.3f}')
print(f'Intercept {slr.intercept_:.3f}')
```

```python
lin_regr_plot(X_std, y_std, slr)
```

# Use of RANSAC(Random Sample Consensus)

```python
from sklearn.linear_model import RANSACRegressor


ransac = RANSACRegressor(LinearRegression(), 
                         max_trials=100, # default
                         min_samples=0.95, 
                         loss='absolute_error', # default
                         residual_threshold=None, # default 
                         random_state=123)


ransac.fit(X_std, y_std)

inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)

line_X = np.arange(3, 10, 1)
line_y_ransac = ransac.predict(line_X[:, np.newaxis])


plt.plot(line_X, line_y_ransac, color='red',ls='-', lw=2,label='Regression')   
plt.scatter(X_std[inlier_mask], y_std[inlier_mask],
            c='steelblue', edgecolor='white', 
            marker='o', label='Inliers')
plt.scatter(X_std[outlier_mask], y_std[outlier_mask],
            c='limegreen', edgecolor='white', 
            marker='s', label='Outliers')

plt.xlabel('Scaled Living area above ground in square feet')
plt.ylabel('Scaled Sale price in U.S. dollars')
plt.legend(loc='upper left')

plt.tight_layout()

plt.show()
```

```python
print(f'Slope: {ransac.estimator_.coef_[0]:.3f}')
print(f'Intercept {ransac.estimator_.intercept_:.3f}')
```

# Evaluating the performance of linear regression models

```python
from sklearn.model_selection import train_test_split

# features (without price)
X = df.iloc[:, :-1].values

# target y(price)
y = df['SalePrice'].values


# split data
X_tr, X_ts, y_tr, y_ts = train_test_split(X,y,  test_size=0.3,
                                         random_state=0)

# model
slr = LinearRegression()
# train
slr.fit(X_tr, y_tr)

# predict
y_tr_pred = slr.predict(X_tr)
y_ts_pred = slr.predict(X_ts)
```

### Residual Scatterplot

```python
x_max = np.max([np.max(y_tr_pred), np.max(y_ts_pred)])
x_min = np.min([np.min(y_tr_pred), np.min(y_ts_pred)])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 3), sharey=True)

ax1.scatter(y_ts_pred, y_ts_pred - y_ts,
            c='limegreen', marker='s', edgecolor='white',
            label='Test data')
ax2.scatter(y_tr_pred, y_tr_pred - y_tr,
            c='steelblue', marker='o', edgecolor='white',
            label='Training data')
ax1.set_ylabel('Residuals')

for ax in (ax1, ax2):
    ax.set_xlabel('Predicted values')
    ax.legend(loc='upper left')
    ax.hlines(y=0, xmin=x_min-100, xmax=x_max+100, color='black', lw=2)

plt.tight_layout()

#plt.savefig('figures/09_11.png', dpi=300)
plt.show()
```

In ideal case the residulas must be 0, or data points should be randomly scattered across the plot. Here one can obseve a repeating pattern which means that regression model doesn't catch all desired dependencies.

### MSE


```python
from sklearn.metrics import mean_squared_error
print(f'Training MSE:{mean_squared_error(y_tr, y_tr_pred):.3f}')
print(f'Test MSE:{mean_squared_error(y_ts, y_ts_pred):.3f}')
```

### MAE

```python
from sklearn.metrics import mean_absolute_error


mae_train = mean_absolute_error(y_tr, y_tr_pred)
mae_test = mean_absolute_error(y_ts, y_ts_pred)
print(f'MAE train: {mae_train:.3f}')
print(f'MAE test: {mae_test:.3f}')
```

### $R^2$

```python
from sklearn.metrics import r2_score
print(f'R^2:{r2_score(y_tr, y_tr_pred):.3f}')
print(f'R^2:{r2_score(y_ts, y_ts_pred):.3f}')

```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
X = np.array([258.0, 270.0, 294.0, 
              320.0, 342.0, 368.0, 
              396.0, 446.0, 480.0, 586.0])\
             [:, np.newaxis]

y = np.array([236.4, 234.4, 252.8, 
              298.6, 314.2, 342.2, 
              360.8, 368.0, 391.2,
              390.8])
# models
lr = LinearRegression()
pr = LinearRegression()

quadratic = PolynomialFeatures(degree=2)
X_quad = quadratic.fit_transform(X)
```

```python
# linear 
lr.fit(X,y)
X_fit = np.arange(250, 600, 10)[:, np.newaxis]
y_lin_fit = lr.predict(X_fit)
```

```python
# polynomial
pr.fit(X_quad, y)
y_quad_fit = pr.predict(quadratic.fit_transform(X_fit))
```

```python
# plot results
plt.scatter(X, y, label='Training points')
plt.plot(X_fit, y_lin_fit, label='Linear fit', linestyle='--')
plt.plot(X_fit, y_quad_fit, label='Quadratic fit')
plt.xlabel('Explanatory variable')
plt.ylabel('Predicted or known target values')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# predict
y_lin_pred = lr.predict(X)
y_quad_pred = pr.predict(X_quad)
```

```python
# Evaluate metrics
mse_lin = mean_squared_error(y, y_lin_pred)
mse_quad = mean_squared_error(y, y_quad_pred)
print(f'Training MSE linear: {mse_lin:.3f}'
      f', quadratic: {mse_quad:.3f}')


r2_lin = r2_score(y, y_lin_pred)
r2_quad = r2_score(y, y_quad_pred)
print(f'Training R^2 linear: {r2_lin:.3f}'
      f', quadratic: {r2_quad:.3f}')
```

Quadratic

Cubic
